BIVARIATE ANALYSIS

Scatter plot

`usage`:to show relationship between two numerical variables and check correlation

numerical vs numerical

On Nitrogen_Level and Yield_Last_Season

```python
#nitrogen level vs yeild last season
sns.scatterplot(x='Nitrogen_Level',y='Yield_Last_Season',data=df)
plt.xlabel('nitrogen level')
plt.ylabel('last season  yeild')
plt.title("impact of the nitrogen level on crop yield")
plt.show()
```

`conclusion`:nitrogen values appear as outliers and nitrogen does not significantly increase yield

On  Rainfall and Yield_Last_Season

```python
#rainfall vs yeild last season
sns.scatterplot(x='Rainfall',y='Yield_Last_Season',data=df)
plt.xlabel('rainfall')
plt.ylabel('last season  yeild')
plt.title('rainfall vs crop yield relationship')
plt.show()
```

`conclusion`:rainfall does not significantly increase yield

On temperature ans humidity

```python
#temperature vs humidity
sns.scatterplot(x='Temperature',y='Humidity',data=df)
plt.xlabel('temperature')
plt.ylabel('humidity')
plt.title('effect of temperature on soil moisture')
plt.show()
```

`conclusion`:most values are clustered in lower temperature range

Line plot

`usage`:to show trend over time and continuous range

On season and yield last season

```python
#season vs yield last season 
season_avg=df.groupby('Season')['Yield_Last_Season'].mean().reset_index()
sns.lineplot(x='Season',y='Yield_Last_Season',data=season_avg,marker='o')
plt.title('average crop yield across different seasons')
```

`conclusion`:this indicates that crop productivity slightly varies across seasons with rabi having compartively lower yield

On Crop_Growth_Stage  and  Nitrogen_Level

```python
#Crop_Growth_Stage vs Nitrogen_Level
avg=df.groupby('Crop_Growth_Stage')['Nitrogen_Level'].mean().reset_index()
sns.lineplot(x='Crop_Growth_Stage',y='Nitrogen_Level',data=avg,marker='o')
plt.title('nitrogen level variation across crop growth stages')
```

`conclusion`:this indicates nitrogen demand increases during acive growth and decreases afterward

Violin plot

`usage`:shows distribution shape and density and better understanding of variation


On Nitrogen_Level and Recommended_Fertilizer

```python
sns.violinplot(x='Nitrogen_Level',y='Recommended_Fertilizer',data=df)
plt.title('nitrogen level vs recommmended fertilizer')
```

`conclusion`:a few fertilizers show slightly higher median nitrogen levels,indicating moderate variation but no extreme differences amoung fertilizer types

Regression plot

`usage`:to show scatter points, regression line and confidence interval

On rainfall and yield last season

```python
plt.figure(figsize=(8,6))
sns.regplot(data=df,x='Rainfall',y='Yield_Last_Season')
plt.title("Regression: Rainfall vs Yield")
plt.xlabel("Rainfall")
plt.ylabel("Yield Last Season")
plt.show()
```

`conclusion`:this indicates that rainfall alone does not strongly determine crop yield

MULTIVARIATE ANALYSIS

Grouped bar chart

`usage`:to compare categorical data and shows counts or averages of the categories

On season and crop type

```python
#season vs crop type
season_crop = pd.crosstab(df['Season'], df['Crop_Type'], normalize='index') * 100
season_crop
```

```python
season_crop.plot(kind='bar', figsize=(10,6))
plt.xlabel("Season")
plt.ylabel("Percentage")
plt.title("Percentage Distribution of Crop Types Across Seasons")
plt.xticks(rotation=0)
plt.legend(title="Crop Type")
plt.show()
```

```python
temp = pd.crosstab(df['Region'], df['Recommended_Fertilizer'], normalize='index') * 100
temp
```

On region and recommended fertilizer

```python
temp.plot(kind='bar', figsize=(10,4))
plt.xlabel("region")
plt.ylabel("Fertilizer")
plt.title("percentage distribution of the fertilizer across all regions")
plt.xticks(rotation='vertical')
plt.legend(loc='upper left',title="fertilizer type")
plt.show()
```

Stacked bar chart


mutli-category comparison


understand the fertilizer demand pattern

```python
stacked_data = pd.crosstab([df['Season'], df['Region']], df['Recommended_Fertilizer'])
```

On season,region and recommended fertilizer

```python
#season,region and recommended fertilizer
stacked_data.plot(kind='bar',stacked=True,figsize=(12,6))
plt.xlabel("Season and Region")
plt.ylabel("Count")
plt.title("Stacked Fertilizer Distribution by Season and Region")
plt.xticks(rotation=45)
plt.show()
```

```python
stacked_data = pd.crosstab([df['Crop_Type'], df['Irrigation_Type']], df['Recommended_Fertilizer'])
stacked_data
```

On crop type,irrigation type and recommended fertilizer

```python
stacked_data.plot(kind='bar',stacked=True,figsize=(12,6))
plt.xlabel("crop type and irrigation")
plt.ylabel("Count")
plt.title("Stacked Fertilizer Distribution by crop type  and irrigation type")
plt.xticks(rotation=45)
plt.show()
```

Area plot

`Usage`:to show cumulative totals over time

On Crop_Growth_Stage,Nitrogen_Level, Phosphorus_Level and Potassium_Level

```python
# Group by growth stage and take mean of Nitrogen Level, Phosphorus Level and Potassium Level
area_data = df.groupby('Crop_Growth_Stage')[['Nitrogen_Level', 'Phosphorus_Level', 'Potassium_Level']].mean()
# Plot stacked area
area_data.plot( kind='area', stacked=True, figsize=(10,6))
plt.xlabel("Crop Growth Stage")
plt.ylabel("Average Nutrient Level")
plt.title("NPK Levels Across Crop Growth Stages")
plt.xticks(rotation=45)
plt.show()
```

Bubble plot

`usage`:to visualize multi-dimensional data clearly

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=df,x='Rainfall',y='Yield_Last_Season',size='Nitrogen_Level',hue='Region',sizes=(20, 400),alpha=0.9)
plt.title("Rainfall vs Yield with Nitrogen Level")
plt.show()
```

`conclusion`:indicating that neither rainfall nor nitrogen levels alone strongly determines the yield and regional differences may also influences productivity
